```python
import model
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

### Image flipping

```python
data = pd.read_csv('driving_log.csv')

idx = 1050
test_img = data['center'][idx].strip()
img = mpimg.imread(test_img)
steering = data['steering'][idx]
```

```python
plt.subplot(1,2,1)
plt.imshow(img);
plt.title('Angle: %.3f' % steering);
plt.axis('off')

img2, steering2 = model.random_horizontal_flip(img, steering)
plt.subplot(1,2,2)
plt.imshow(img2);
plt.title('Angle: %.3f' % steering2);
plt.axis('off')
```

```python
plt.subplot(1,2,1)
plt.imshow(img);
plt.title('Angle: %.3f' % steering);
plt.axis('off')

img2, steering2 = model.random_translation(img, steering)
plt.subplot(1,2,2)
plt.imshow(img2);
plt.title('Angle: %.3f' % steering2);
plt.axis('off')
```

```python
img_l = mpimg.imread(data['left'][idx].strip())
img_r = mpimg.imread(data['right'][idx].strip())

plt.subplot(1,3,1)
plt.imshow(img_l);
plt.title('Angle: %.3f' % (steering + model.ANGLE_OFFSET));
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img);
plt.title('Angle: %.3f' % steering);
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_r);
plt.title('Angle: %.3f' % (steering - model.ANGLE_OFFSET));
plt.axis('off')
```
